```python
from math import sin, cos, sqrt, atan2, radians

def distance(lat1, lon1, lat2, lon2):
    
    # approximate radius of earth in km
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
        
    distance = R * c
    return distance
```

```python
df_ticket['distance'] = df_ticket.apply(lambda x: distance(x.o_latitude, x.o_longitude, x.d_latitude, x.d_longitude), axis=1)
```

#### on a ajouté une colone 'distance' a dataframe ticktes ou on a calculé la distance entre les cities de depart et de l'arrivée

```python
def categorie_distance(distance):
    
    if (distance < 201):
        categorie = 'A'
    elif (distance >=201) & (distance < 800):
        categorie = 'B'
    elif (distance >=800) & (distance < 2000):
        categorie = 'C'
    else :
        categorie='D'
     
    return categorie
```

```python
df_ticket['categorie'] = df_ticket.apply(lambda x: categorie_distance(x.distance), axis=1)
```

#### on remarque qui'il n y a pas la catégorie 'D' , c'est à dire il n y a pas une distance plus de 2000km
